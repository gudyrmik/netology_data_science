```python
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(A[:, 0], A[:, 1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
plt.plot(A[:, 0] / 2, A[:, 1] / 2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
th = np.radians(130)
R = np.array(( (np.cos(th), -np.sin(th)),
               (np.sin(th),  np.cos(th)) ))
A_rotated = R @ np.array([A[:, 0].T, A[:, 1].T])
plt.plot(A_rotated[0], A_rotated[1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
T = np.array(( (1, 1),
               (1, -1) ))
A_flipped = T @ np.array([A[:, 0].T, A[:, 1].T])
plt.plot(A_flipped[0], A_flipped[1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
np.array([A[:, 0].T, A[:, 1].T])
```
